```python
import matplotlib.pyplot as plt


total_sales_city = df.groupby('City')['Quantity'].sum()

# Plotting
plt.figure(figsize=(8, 4))
bar_plot = total_sales_city.plot(kind='bar')
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)

# Adding text annotations
for index, value in enumerate(total_sales_city):
    plt.text(index, value + 0.1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()

# This plot gives the total sales vs city and we got to know sales are equal across as the cities
```

```python

total_sales_branch = df.groupby('Branch')['Quantity'].sum()

# Plotting
plt.figure(figsize=(8, 4))
bar_plot = total_sales_branch.plot(kind='bar')
plt.title('Total Sales by Branch')
plt.xlabel('Branch')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)

# Adding text annotations
for index, value in enumerate(total_sales_branch):
    plt.text(index, value + 0.1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()

# This plot gives the total sales vs branch through this we got to know c branch sales is less compare to other branches
```

```python
df['Revenue'] = df['Unit price'] * df['Quantity']

# Group by city and calculate total revenue
total_revenue_city = df.groupby('City')['Revenue'].sum()
# We have used the formula revenue = unit price * quantity


# Plotting
plt.figure(figsize=(8, 4))
bar_plot = total_revenue_city.plot(kind='bar')
plt.title('Revenue by City')
plt.xlabel('City')
plt.ylabel('Revenue')
plt.xticks(rotation=45)

# Adding text annotations
for index, value in enumerate(total_revenue_city):
    plt.text(index, value + 0.1, str(round(value, 2)), ha='center', va='bottom')

plt.tight_layout()
plt.show()

# This plot gives the total revenue vs city and we got to know revenue are equal across as the cities
```

```python
total_revenue_branch = df.groupby('Branch')['Revenue'].sum()

# Plotting
plt.figure(figsize=(8, 4))
bar_plot = total_revenue_branch.plot(kind='bar')
plt.title('Revenue by Branch')
plt.xlabel('Branch')
plt.ylabel('Revenue')
plt.xticks(rotation=0)

# Adding text annotations
for index, value in enumerate(total_revenue_branch):
    plt.text(index, value + 0.1, str(round(value, 2)), ha='center', va='bottom')

plt.tight_layout()
plt.show()

# This plot gives the total revenue vs branch through this we got to know c branch revenue is less compare to other branches
```

```python
import seaborn as sns

# Group by both city and branch and calculate total sales
sales_data = df.groupby(['City', 'Branch'])['Quantity'].sum().reset_index()
print(sales_data)

# Plotting
plt.figure(figsize=(8, 4))
sns.barplot(data=sales_data, x='City', y='Quantity', hue='Branch')
plt.title('Sales by City and Branch')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# This plot is giving the total sales of each city w.r.t the branch we can se that C branch has poor performance in Mandalay City while other cities branches have shown equal performance
```

```python
revenue_data = df.groupby(['City', 'Branch'])['Revenue'].sum().reset_index()
print(revenue_data)

# Plotting
plt.figure(figsize=(8, 4))
sns.barplot(data=revenue_data, x='City', y='Revenue', hue='Branch')
plt.title('Revenue by City and Branch')
plt.xlabel('City')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# This plot is giving the total revenue of each city w.r.t the branch we can se that C branch has poor performance in Mandalay City while other cities branches have shown equal performance

```

```python

df['Date'] = pd.to_datetime(df['Date'])


df['Month'] = df['Date'].dt.month

df['Revenue'] = df['Quantity'] * df['Unit price']


monthly_data = df.groupby(['Month', 'Product line']).agg({'Quantity': 'sum', 'Revenue': 'sum'}).reset_index()

print(monthly_data)
```

```python
import seaborn as sns


# Plotting
plt.figure(figsize=(8, 4))
sns.barplot(data=monthly_data, x='Month', y='Quantity', hue='Product line')
plt.title('Monthly Sales by Product Line')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Product Line',  bbox_to_anchor=(1, 1), loc='upper left')
plt.show()

# Sales of our product were least in 2nd february dip in the sales of fashion accesories need to focus be focused for April so that we can improve our sales
```

```python

monthly_gender_data = df.groupby(['Month', 'Gender']).agg({'Quantity': 'sum', 'Revenue': 'sum'}).reset_index()


print(monthly_gender_data)
```

```python


# Plotting
plt.figure(figsize=(8, 4))
sns.barplot(data=monthly_gender_data, x='Month', y='Quantity', hue='Gender')
plt.title('Monthly Sales by Gender')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Gender')
plt.show()
```

```python

monthly_payment_count = df.groupby(['Month', 'Payment']).size().reset_index(name='Sales')

print(monthly_payment_count)
```
